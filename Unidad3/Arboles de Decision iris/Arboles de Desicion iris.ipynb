```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn import tree

iris= load_iris()
iris
```

```python
clf = tree.DecisionTreeClassifier()

#Primer parametro son los valores de las medidas de las flores
#El segundo parametro es la varibale que se queire predecir
clf = clf.fit(iris.data, iris.target )

clf.score(iris.data, iris.target)

#Genera la salida en formato .dot
#y lo podemos visualizar el archivo DOT con la libreria graphviz
tree.plot_tree(clf)
```

```python
#Generar una imagen
import graphviz
dot_data = tree.export_graphviz(clf, out_file=None)
graph = graphviz.Source(dot_data)
graph.render('iris', format='png')
```
